```python
import pandas as pd
```

```python
# Load the dataset
data_path = '/Users/nathanyap/Desktop/DataMining_Project/project/Nathan Findings/pro_admits_cleaned_data.csv'
df_admitsFYI = pd.read_csv(data_path)
```

```python
df_admitsFYI
```

```python
# Convert TOEFL scores to a scale with a maximum of 10
# will be using 120 as the max score for TOEFL
df_admitsFYI['TOEFL_Scaled'] = (df_admitsFYI['TOEFL'] / 120) * 10
df_admitsFYI['TOEFL_Scaled'] = df_admitsFYI['TOEFL_Scaled'].round(1)
```

```python
# Now we also want to convert the IELTS score to a scale of Max 10
# We will be using 9 as the max score for IELTS
df_admitsFYI['IELTS_Scaled'] = (df_admitsFYI['IELTS'] / 9) * 10
df_admitsFYI['IELTS_Scaled'] = df_admitsFYI['IELTS_Scaled'].round(1)
```

```python
# df['Combined_Score'] = df.apply(lambda row: row['IELTS_Scaled'] if row['TOEFL_Scaled'] == 0.0 else (row['TOEFL_Scaled'] if row['IELTS_Scaled'] == 0.0 else 0.0 if row['TOEFL_Scaled'] == 0.0 and row['IELTS_Scaled'] == 0.0 else row['TOEFL_Scaled']), axis=1)
```

```python
def get_combined_score(row):
    # If TOEFL_Scaled is not 0, use it
    if row['TOEFL_Scaled'] != 0.0:
        return row['TOEFL_Scaled']
    # If IELTS_Scaled is not 0, use it
    elif row['IELTS_Scaled'] != 0.0:
        return row['IELTS_Scaled']
    # If both are 0, return 0
    else:
        return 0.0

# Apply the function to each row
df_admitsFYI['Combined_Score'] = df_admitsFYI.apply(get_combined_score, axis=1)
```

```python
df_admitsFYI
```

```python
df_admitsFYI.drop(columns=['TOEFL', 'IELTS', 'TOEFL_Scaled', 'IELTS_Scaled'], inplace=True)
```

```python
df_admitsFYI.rename(columns={'Combined_Score': 'TOEFL/IELTS'}, inplace=True)
```
