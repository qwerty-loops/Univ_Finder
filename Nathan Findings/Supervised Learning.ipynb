# Decision Trees with OPTION 1 - 5 inputs

# TOEFL

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
```

```python
X = df_admitsFYI[['GPA', 'GRE Total', 'TOEFL', 'Work Exp', 'Papers']].copy()

# Converting 'Work Exp' to a numerical format if it's not already
# Make sure 'Work Exp' is in a suitable numeric format for modeling
X['Work Exp'] = pd.to_numeric(X['Work Exp'], errors='coerce').fillna(0)

y = df_admitsFYI['Status']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2%}")
print(f"Model Precision: {precision:.2%}")
print(f"Model Recall: {recall:.2%}")
```

# RANDOM FOREST for TOELF AND IELTS

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Assuming you have already prepared X and y, and split them into training and test sets
X_TOEFL = df_admitsFYI[['GPA', 'GRE Total', 'TOEFL', 'Work Exp', 'Papers']].copy()
X_IELTS = df_admitsFYI[['GPA', 'GRE Total', 'IELTS', 'Work Exp', 'Papers']].copy()

y = df_admitsFYI['Status']

# Split the data into training and test sets
X_train_TOEFL, X_test_TOEFL, y_train_TOEFL, y_test_TOEFL = train_test_split(X_TOEFL, y, test_size=0.3, random_state=42)
X_train_IELTS, X_test_IELTS, y_train_IELTS, y_test_IELTS = train_test_split(X_IELTS, y, test_size=0.3, random_state=42)

# Initialize the Random Forest classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
rf_clf.fit(X_train_TOEFL, y_train_TOEFL)
rf_clf.fit(X_train_IELTS, y_train_IELTS)

# Predict on the test set
y_pred_rf_TOEFL = rf_clf.predict(X_test_TOEFL)
y_pred_rf_IELTS = rf_clf.predict(X_test_IELTS)

# Evaluate the model TOEFL
accuracy_rf_TOEFL = accuracy_score(y_test_TOEFL, y_pred_rf_TOEFL)
precision_rf_TOEFL = precision_score(y_test_TOEFL, y_pred_rf_TOEFL)
recall_rf_TOEFL = recall_score(y_test_TOEFL, y_pred_rf_TOEFL)

# Evaluate the model IELTS
accuracy_rf_IELTS = accuracy_score(y_test_IELTS, y_pred_rf_IELTS)
precision_rf_IELTS = precision_score(y_test_IELTS, y_pred_rf_IELTS)
recall_rf_IELTS = recall_score(y_test_IELTS, y_pred_rf_IELTS)



# Print the scores
print(f"Random Forest Model Accuracy for TOEFL: {accuracy_rf_TOEFL:.2%}")
print(f"Random Forest Model Precision for TOEFL: {precision_rf_TOEFL:.2%}")
print(f"Random Forest Model Recall for TOEFL: {recall_rf_TOEFL:.2%}")

print(f"Random Forest Model Accuracy for IELTS: {accuracy_rf_IELTS:.2%}")
print(f"Random Forest Model Precision for IELTS: {precision_rf_IELTS:.2%}")
print(f"Random Forest Model Recall for IELTS: {recall_rf_IELTS:.2%}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined

X_TOEFL = df_admitsFYI[['GPA', 'GRE Total', 'TOEFL', 'Work Exp', 'Papers']].copy()
X_IELTS = df_admitsFYI[['GPA', 'GRE Total', 'IELTS', 'Work Exp', 'Papers']].copy()

y = df_admitsFYI['Status']

# Split the data
X_train_TOEFL, X_test_TOEFL, y_train_TOEFL, y_test_TOEFL = train_test_split(X_TOEFL, y, test_size=0.3, random_state=42)
X_train_IELTS, X_test_IELTS, y_train_IELTS, y_test_IELTS = train_test_split(X_IELTS, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled_TOEFL = scaler.fit_transform(X_train_TOEFL)
X_test_scaled_TOEFL = scaler.transform(X_test_TOEFL)

scaler = StandardScaler()
X_train_scaled_IELTS = scaler.fit_transform(X_train_IELTS)
X_test_scaled_IELTS = scaler.transform(X_test_IELTS)

# Initialize the MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, activation='relu', solver='adam', random_state=42)

# Train the model
mlp_clf.fit(X_train_scaled_TOEFL, y_train_TOEFL)
mlp_clf.fit(X_train_scaled_IELTS, y_train_IELTS)

# Predict on the test set
y_pred_nn_TOEFL = mlp_clf.predict(X_test_scaled_TOEFL)
y_pred_nn_IELTS = mlp_clf.predict(X_test_scaled_IELTS)

# Evaluate the model TOEFL
accuracy_nn_TOEFL = accuracy_score(y_test, y_pred_nn_TOEFL)
precision_nn_TOEFL = precision_score(y_test, y_pred_nn_TOEFL)
recall_nn_TOEFL = recall_score(y_test, y_pred_nn_TOEFL)

# Evaluate the model IELTS
accuracy_nn_IELTS = accuracy_score(y_test, y_pred_nn_IELTS)
precision_nn_IELTS = precision_score(y_test, y_pred_nn_IELTS)
recall_nn_IELTS = recall_score(y_test, y_pred_nn_IELTS)

# Print the scores TOEFL
print(f"Neural Network Model Accuracy: {accuracy_nn_TOEFL:.2%}")
print(f"Neural Network Model Precision: {precision_nn_TOEFL:.2%}")
print(f"Neural Network Model Recall: {recall_nn_TOEFL:.2%}")
print()

# Print the scores TOEFL
print(f"Neural Network Model Accuracy: {accuracy_nn_IELTS:.2%}")
print(f"Neural Network Model Precision: {precision_nn_IELTS:.2%}")
print(f"Neural Network Model Recall: {recall_nn_IELTS:.2%}")
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

X_TOEFL = df_admitsFYI[['GPA', 'GRE Total', 'TOEFL', 'Work Exp', 'Papers']].copy()
X_IELTS = df_admitsFYI[['GPA', 'GRE Total', 'IELTS', 'Work Exp', 'Papers']].copy()

y = df_admitsFYI['Status']

# Split the data
X_train_TOEFL, X_test_TOEFL, y_train_TOEFL, y_test_TOEFL = train_test_split(X_TOEFL, y, test_size=0.3, random_state=42)
X_train_IELTS, X_test_IELTS, y_train_IELTS, y_test_IELTS = train_test_split(X_IELTS, y, test_size=0.3, random_state=42)

# Initialize the Gradient Boosting classifier
gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the model on the training set
gb_clf.fit(X_train_scaled_TOEFL, y_train_TOEFL)
gb_clf.fit(X_train_scaled_IELTS, y_train_IELTS)                             

# Predict on the test set
y_pred_gb_TOEFL = gb_clf.predict(X_test_TOEFL)
y_pred_gb_IELTS = gb_clf.predict(X_test_IELTS)                                                                           

# Evaluate the model for TOEFL
accuracy_gb_TOEFL = accuracy_score(y_test_TOEFL, y_pred_gb_TOEFL)
precision_gb_TOEFL = precision_score(y_test_TOEFL, y_pred_gb_TOEFL)
recall_gb_TOEFL = recall_score(y_test_TOEFL, y_pred_gb_TOEFL)
                                                                            
# Evaluate the model for TOEFL
accuracy_gb_IELTS = accuracy_score(y_test_IELTS, y_pred_gb_IELTS)
precision_gb_IELTS = precision_score(y_test_IELTS, y_pred_gb_IELTS)
recall_gb_IELTS = recall_score(y_test_IELTS, y_pred_gb_IELTS)

# Print the scores for TOEFL
print(f"Gradient Boosting Model Accuracy for TOEFL: {accuracy_gb_TOEFL:.2%}")
print(f"Gradient Boosting Model Precision for TOEFL: {precision_gb_TOEFL:.2%}")
print(f"Gradient Boosting Model Recall for TOEFL: {recall_gb_TOEFL:.2%}")
print()
# Print the scores for IELTS
print(f"Gradient Boosting Model Accuracy for IELTS: {accuracy_gb_IELTS:.2%}")
print(f"Gradient Boosting Model Precision for IELTS: {precision_gb_IELTS:.2%}")
print(f"Gradient Boosting Model Recall for IELTS: {recall_gb_IELTS:.2%}")
```
